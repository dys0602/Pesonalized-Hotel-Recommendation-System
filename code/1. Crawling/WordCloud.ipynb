# noun_extraction column 전처리 + 형태 변환

### 모든 특수문자, 마침표, 쉼표, 숫자, 한글자 씩 인덱스화 된 것을 전처리하기

```python
w_list = []

for i in range(0, 31):
  # 전처리
  s = copy_df['noun_extraction'][i].replace("(", "")
  s = s.replace(")", "")
  s = s.replace("[", "")
  s = s.replace("]", "")
  s = s.replace("'", "")
  s = s.replace(" ", "")
  # 쉼표를 중심으로 문자열 단어와 문자열 숫자 나누기
  s = s.split(',')
  # 짝수 인덱스 -> 단어  |  홀수 인덱스 -> 빈도로 나누기
  word = s[0::2]
  freq = s[1::2]
  # 문자열 숫자를 실수형으로 바꾸기
  freq = list(map(int,freq))
  # 모두 빈 리스트에 넣어주기
  w_list.append(word)
  w_list.append(freq)
```

### w_list에서 짝수 인덱스는 단어 리스트에, 홀수 인덱스는 빈도 리스트에 저장

```python
# 단어와 빈도 각각 나눠서 변수 지정
word_list = w_list[0::2]
freq_list = w_list[1::2]
```

### 워드클라우드 생성 시, 딕셔너리만 사용 가능하기 때문에 단어-빈도의 쌍을 딕셔너리 형태로 변환

```python
# for loop으로 새로운 딕셔너리 30개 생성
dic = {}

for i in range(0, 31):
  dic[i] = {}
```

```python
for i in tqdm(range(0, 31)):
  # 31개 빈 딕셔너리에 합친 두 리스트 key: value로 저장
  dic[i] = {name:value for name, value in zip(word_list[i], freq_list[i])}
```

```python
# 구성비 -> 가성비로 교체
for i in range(0, 31):
  if '구성비' in dic[i]:
    dic[i]['가성비'] = dic[i].pop('구성비')
```

```python
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt   
%matplotlib inline
from IPython.display import set_matplotlib_formats
matplotlib.rc('font',family = 'NanumGothic')
set_matplotlib_formats('retina')
matplotlib.rc('axes',unicode_minus = False)

# 호텔별로 워드클라우드 생성 
for i in tqdm(range(0, 31)):
  wc = WordCloud(font_path ='/usr/share/fonts/truetype/nanum/NanumGothic.ttf', background_color='white',
                 random_state=None, colormap = "Accent_r", width=1500, height=1000).generate_from_frequencies(dic[i])
  plt.figure(1, figsize=(15,10))
  plt.imshow(wc)
  plt.axis('off')
  plt.show()
```

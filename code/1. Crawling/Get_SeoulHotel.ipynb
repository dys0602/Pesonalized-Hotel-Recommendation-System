# Get jeju hotel info
- hotel code, hotel name, location info

### 제주 호텔 불러오는 함수

```python
# def 
def hotel_info(txt_url):
  global hotel_info_df
  # 호텔 url, name을 담기 위한 리스트 생성
  hotel_code = []
  hotel_name = []
  hotel_location = []

  # 반복문을 통해 호텔 url, name 담기
  txt = pd.read_table(txt_url)
  txt = txt.to_string() #dataframe을 string으로 변환
  hotel_info = txt.split("href=") # href 뒤에 호텔별 url 주소가 나오므로 먼저 'href=' 기준으로 split

  for i in range(1, len(hotel_info)): # 0번째는 사용할 필요 없으므로 1번째부터
    # 호텔 code
    url = hotel_info[i].split()[0].replace('"', "") # 필요없는 '"' 부분 제거
    hotel_code.append(url.split('hotel/')[1])

    # 호텔명
    name = hotel_info[i].split("title=") # 'title=' 뒤에 호텔명 나오므로 split
    name = name[1].split(">")[0].replace('"', "") # '>'까지 호텔명 받고 나머지 제거([0]) 후, '"' 부분 제거
    hotel_name.append(name)

  for i in range(0, len(hotel_name)):
    hotel_name[i] = hotel_name[i].replace(" class=common_clearfix__M6urU", "") # hotel_name 중에 일부 문자열이 생성되므로 삭제 

  hotel_info_df = pd.DataFrame(x for x in zip(hotel_code, hotel_name)) # 두가지 list 엮어서 dataframe화
  hotel_info_df.columns = ['hotel_code', 'hotel_name']
  return hotel_info_df
```
